```python
df['all_crimes'] = df['violent_crimes'] + df['homicides'] + df['rapes'] + df['assaults'] + df['robberies']

df['all_crimes_percapita'] = df['all_crimes']/(df['population']/100000)
```

```python
df['all_crimes'].head(10)
```

```python
df['all_crimes_percapita'].head(10)
```

Now to find the city with maximum crimes per capita, we will have to create a new column, 'all_crimes', by adding up all different types of crimes given in the dataset. Then we will divide this column by the city population to get count of crimes per capita. This will enable to priortise and target cities in particular that needs more attention. 

```python
df[df['report_year'] == 2015][['agency_jurisdiction', 'all_crimes_percapita']].sort_values(by=['all_crimes_percapita'], ascending=False)
```

Here we can see the city-wise per capita crimes count for the year 2015 that is the current condition. AS per this St. Louis City at Missouri has the maximum crime cases per capita in 2015. We will focus our further analysis over this city.

## 6. Plot

We will use the matplotlib library to plot a line chart of assaults reported, per capita on yearly basis.

```python
plt.plot(stlouis_df['report_year'], stlouis_df['assaults_percapita'])

plt.scatter(stlouis_df['report_year'], stlouis_df['assaults_percapita'])

plt.title("Assaults reported (per capita) between 1975 to 2015")

plt.show()
```

The plot above shows count of assaults reported per 100,000 people. We can see that the number of assaults cases were relatively low between 1975 and 1980, but then such cases beacame more frequent and the situation was quite serious between late eighties and mid-nineties.Since 1995, we can see a downward trend in the number of assault cases.
